```python
# ============================================================
# Floorplan Label + SqFt Extraction (Main rooms + Subrooms AA..AG, A4, etc.)
# Robust OCR pipeline for floorplans using Tesseract:
#  - strong preprocessing + text thickening
#  - dense tiling (6x6)
#  - rotations (0, 90, 270) to catch vertical labels
#  - word-level OCR (image_to_data) + normal OCR text
#  - strict-ish parsing to avoid "46" from "464sf" errors
# Outputs:
#  - outputs/tesseract_all_text.txt
#  - outputs/floorplan_label_sqft_pairs.csv
# ============================================================

!apt-get -qq update
!apt-get -qq install -y tesseract-ocr
!pip -q install pytesseract gdown pandas pillow numpy

from pathlib import Path
import re
import numpy as np
import pandas as pd
import gdown
from PIL import Image, ImageOps, ImageEnhance, ImageFilter
import pytesseract

# ----------------------------
# Paths / download
# ----------------------------
data_dir = Path("data"); data_dir.mkdir(exist_ok=True)
out_dir  = Path("outputs"); out_dir.mkdir(exist_ok=True)

DRIVE_FILE_ID = "1xOF3SYM1EN2RxgUgo9j4ycCwGFYmzOJv"
img_path = data_dir / "floorplan_big.png"

if not img_path.exists():
    gdown.download(f"https://drive.google.com/uc?id={DRIVE_FILE_ID}", str(img_path), quiet=False)

img = Image.open(img_path).convert("RGB")
print("Loaded:", img_path, "| size:", img.size)

# ----------------------------
# Preprocessing utilities
# ----------------------------
def otsu_threshold(gray_np):
    hist = np.bincount(gray_np.ravel(), minlength=256).astype(np.float64)
    total = gray_np.size
    sum_total = np.dot(np.arange(256), hist)

    sumB = 0.0
    wB = 0.0
    max_var = -1.0
    thr = 127

    for t in range(256):
        wB += hist[t]
        if wB == 0:
            continue
        wF = total - wB
        if wF == 0:
            break
        sumB += t * hist[t]
        mB = sumB / wB
        mF = (sum_total - sumB) / wF
        var_between = wB * wF * (mB - mF) ** 2
        if var_between > max_var:
            max_var = var_between
            thr = t
    return thr

def preprocess_floorplan(pil_img, upscale=6):
    """
    Strong floorplan preprocessing:
    - upscale
    - grayscale
    - boost contrast/sharpness
    - otsu binarize
    - invert if needed
    - thicken strokes (min/max filter combo)
    """
    w, h = pil_img.size
    pil = pil_img.resize((w*upscale, h*upscale), Image.LANCZOS)

    gray = ImageOps.grayscale(pil)
    gray = ImageEnhance.Contrast(gray).enhance(2.8)
    gray = ImageEnhance.Sharpness(gray).enhance(2.4)
    gray = gray.filter(ImageFilter.MedianFilter(size=3))

    g = np.array(gray, dtype=np.uint8)
    thr = otsu_threshold(g)
    bw = (g > thr).astype(np.uint8) * 255

    # invert if mostly black background
    if bw.mean() < 127:
        bw = 255 - bw

    bw_img = Image.fromarray(bw)

    # Thicken strokes: MIN then MAX helps connect broken letters
    bw_img = bw_img.filter(ImageFilter.MinFilter(size=3))
    bw_img = bw_img.filter(ImageFilter.MaxFilter(size=3))

    return bw_img

def make_tiles(pil_img, rows=6, cols=6, overlap=0.18):
    w, h = pil_img.size
    ox = int(w * overlap)
    oy = int(h * overlap)
    tiles = []
    for r in range(rows):
        for c in range(cols):
            x0 = int(c * w / cols) - (ox if c > 0 else 0)
            y0 = int(r * h / rows) - (oy if r > 0 else 0)
            x1 = int((c + 1) * w / cols) + (ox if c < cols - 1 else 0)
            y1 = int((r + 1) * h / rows) + (oy if r < rows - 1 else 0)
            x0, y0 = max(0, x0), max(0, y0)
            x1, y1 = min(w, x1), min(h, y1)
            tiles.append(pil_img.crop((x0, y0, x1, y1)))
    return tiles

# ----------------------------
# OCR (text + word-level)
# ----------------------------
def tesseract_cfg():
    # allow only typical floorplan chars; helps reduce junk
    whitelist = "ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyz0123456789-[]"
    return f'-l eng --oem 1 -c tessedit_char_whitelist={whitelist}'

def ocr_text(pil_img, psm):
    cfg = tesseract_cfg() + f" --psm {psm}"
    return pytesseract.image_to_string(pil_img, config=cfg) or ""

def ocr_words(pil_img, psm):
    """
    Word-level OCR table with confidence.
    """
    cfg = tesseract_cfg() + f" --psm {psm}"
    df = pytesseract.image_to_data(pil_img, config=cfg, output_type=pytesseract.Output.DATAFRAME)
    df = df.dropna(subset=["text"])
    df["text"] = df["text"].astype(str)
    return df

def ocr_tile_all(tile):

    texts = []
    words_all = []

    for rot in (0, 90, 270):
        timg = tile if rot == 0 else tile.rotate(rot, expand=True)

        # free-text OCR
        for psm in (6, 11, 12):
            texts.append(ocr_text(timg, psm))

        # word-level OCR (useful when full text is messy)
        for psm in (6, 11):
            wdf = ocr_words(timg, psm)
            # keep reasonable confidence only (tune if needed)
            wdf = wdf[wdf["conf"] >= 30]
            words_all.append(wdf)

    words_df = pd.concat(words_all, ignore_index=True) if words_all else pd.DataFrame()
    return "\n".join(texts), words_df

# ----------------------------
# Parsing helpers
# ----------------------------
def norm(s: str) -> str:
    s = s.upper()
    s = s.replace("—", "-").replace("–", "-")
    # common OCR mistake: 5F instead of SF
    s = s.replace("5F", "SF").replace("S F", "SF")
    # normalize brackets spacing
    s = re.sub(r"\s+", " ", s)
    return s.strip()

def extract_pairs_from_text(big_text: str):
    """
    Extract pairs from a big OCR text blob.
    Accept:
      - AL...-... + number + SF
      - [AL...-...] + number + SF
      - subrooms like AA, AB, ... AG + number + SF
      - A4 + number + SF
    """
    t = norm(big_text)

    # We require an SF-like suffix to avoid matching "46" from "464SF".
    sf = r"(?:SF|S F|5F)"

    pairs = []

    # Main rooms (AL...)
    main_label = r"\bAL[0-9A-Z]{0,4}-[0-9A-Z]{1,6}\b"
    for lab, num in re.findall(rf"({main_label})\s*([0-9]{{1,5}})\s*{sf}\b", t):
        pairs.append((lab, int(num)))

    # Bracketed labels
    for lab, num in re.findall(rf"\[([A-Z0-9-]+)\]\s*([0-9]{{1,5}})\s*{sf}\b", t):
        pairs.append((lab, int(num)))

    # Subrooms: AA..AZ (but in your plan it’s AA..AG) and A4-type
    # AA / AB / AC ...
    for lab, num in re.findall(rf"\b(A[A-Z])\s*([0-9]{{1,5}})\s*{sf}\b", t):
        pairs.append((lab, int(num)))

    # A4 type
    for lab, num in re.findall(rf"\b(A\d)\s*([0-9]{{1,5}})\s*{sf}\b", t):
        pairs.append((lab, int(num)))

    return pairs

def extract_pairs_from_words(words_df: pd.DataFrame):
    """
    Extract pairs from word tokens by scanning sequences like:
      LABEL  NUMBER  SF
    This often recovers pairs missed in raw text.
    """
    if words_df is None or words_df.empty:
        return []

    toks = [norm(x) for x in words_df["text"].tolist() if str(x).strip()]
    pairs = []

    def is_main_label(x):
        return re.fullmatch(r"AL[0-9A-Z]{0,4}-[0-9A-Z]{1,6}", x) is not None or re.fullmatch(r"\[AL[0-9A-Z-]+\]", x) is not None

    def is_sub_label(x):
        return re.fullmatch(r"A[A-Z]", x) is not None or re.fullmatch(r"A\d", x) is not None

    def is_num(x):
        return re.fullmatch(r"\d{1,5}", x) is not None

    def is_sf(x):
        return x in {"SF", "S F", "5F"}

    for i in range(len(toks) - 2):
        a, b, c = toks[i], toks[i+1], toks[i+2]

        # normalize bracket token like [AL1-OOPA] -> AL1-OOPA
        a2 = a
        if a2.startswith("[") and a2.endswith("]"):
            a2 = a2[1:-1]

        if (is_main_label(a) or is_main_label(a2) or re.fullmatch(r"AL[0-9A-Z]{0,4}-[0-9A-Z]{1,6}", a2)) and is_num(b) and is_sf(c):
            pairs.append((a2, int(b)))

        if is_sub_label(a) and is_num(b) and is_sf(c):
            pairs.append((a, int(b)))

    return pairs

def finalize_pairs(pairs):
    if not pairs:
        return pd.DataFrame(columns=["label", "sqft", "occurrence"])

    df = pd.DataFrame(pairs, columns=["label", "sqft"])

    # drop obvious junk: sqft 0, or absurdly large
    df = df[(df["sqft"] > 0) & (df["sqft"] < 20000)]

    # keep duplicates (AA can appear multiple times); add occurrence index
    df["occurrence"] = df.groupby("label").cumcount() + 1

    # sort
    df = df.sort_values(["label", "occurrence", "sqft"]).reset_index(drop=True)
    return df

# ============================================================
# Run
# ============================================================

prep = preprocess_floorplan(img, upscale=6)
prep_path = out_dir / "floorplan_preprocessed_strong.png"
prep.save(prep_path)
print("Saved preprocessed:", prep_path, "| size:", prep.size)

tiles = make_tiles(prep, rows=6, cols=6, overlap=0.18)
print("Tiles:", len(tiles), "| tile size ~", tiles[0].size)

all_text = []
all_pairs = []

for i, tile in enumerate(tiles, 1):
    txt, wdf = ocr_tile_all(tile)
    all_text.append(f"\n--- TILE {i} ---\n{txt}\n")

    # parse pairs from both sources
    all_pairs.extend(extract_pairs_from_text(txt))
    all_pairs.extend(extract_pairs_from_words(wdf))

merged_text = "\n".join(all_text)
txt_path = out_dir / "tesseract_all_text.txt"
txt_path.write_text(merged_text, encoding="utf-8")
print("Saved OCR text:", txt_path)

df = finalize_pairs(all_pairs)

if df.empty:
    raise RuntimeError(
        "Still no pairs extracted.\n"
        "Open outputs/tesseract_all_text.txt and check what OCR is reading.\n"
        "Then we’ll tune patterns based on the exact OCR output."
    )

# Optional: de-dup exact duplicates (same label+sqft repeats a lot due to rotations/PSMs)
df_dedup = df.drop_duplicates(subset=["label", "sqft"]).reset_index(drop=True)

print("Total pairs (with repeats):", len(df))
print("Unique label+sqft pairs:", len(df_dedup))
display(df_dedup)

csv_path = out_dir / "floorplan_label_sqft_pairs.csv"
df_dedup.to_csv(csv_path, index=False)
print("Saved CSV:", csv_path)
```
